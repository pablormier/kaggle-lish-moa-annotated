```python
import pandas as pd

# Read dataset directly from the Github repo
df = pd.read_csv('https://github.com/pablormier/kaggle-lish-moa-annotated/raw/master/lish_moa_annotated.csv.zip')
df
```

```python
# Get only the training data (rows for which MoAs and drug ids are known)
df_training = df.loc[df.training==True,]
df_training.shape
```

```python
# Get the annotated gene names. The gene features (originally called g-XX) were replaced by g-GeneSymbol_ENTREZID
genes = [e for e in df_training.columns.values if e.startswith("g-")]
genes[:10]
```

```python
# Get the annotated cell lines. The cell viability features (originally called c-XX) were replaced by c-XX-CCLEcellName.
# The number following the 'c-' corresponds to the original number used in the kaggle dataset.
# Cell lines can be queried using the DepMap portal. For example
# https://depmap.org/portal/cell_line/HEYA8_OVARY?tab=mutation

cells = [e for e in df_training.columns.values if e.startswith("c-")]
cells[:10]
```

```python
# Split the training data into features and labels
X = df_training.iloc[:, :-608]
Y = df_training.iloc[:, -608:]
X.shape, Y.shape
```

```python
# The target labels used for evaluation in the Kaggle competition are only the first 206 labels
Y_target = Y.iloc[:,:206]
Y_target
```
